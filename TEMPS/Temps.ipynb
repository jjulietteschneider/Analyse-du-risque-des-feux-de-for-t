Au niveau de l'espace, on se positionne sur les malles SAFRAN, et on va devoir se reposer (par manque de précision sur les données) sur le lieu de départ du feu, et de l'IFM correspondant à la maille entière. Ce que l'on pourrait regarder néanmoins, c'est l'impact qu'un feu dans une certaine maille pourrait avoir sur l'apparition d'autres feux à proximité (donc soit dans la même maille, soit les mailles à proximité), mais risque
