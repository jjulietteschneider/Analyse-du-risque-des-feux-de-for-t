```python
nbfeuparan = []
for i in range(2006, 2022) :
    nbfeu = df[df['Année'] == i].count().iloc[0]
    nbfeuparan.append(nbfeu)
    print(nbfeu,' en ',i) 

def calculate_mean(numbers):
    mean = sum(numbers) / len(numbers)
    return mean

calculate_mean(nbfeuparan)
```

On trouve une moyenne de 2.6k feux par an entre 2006 et 2021, ce qui est bien loin des 3.5k feux par an trouvés dans la notice de Generali.

```python
def assign_value(x):
    if x < (1000*10000):
        return 0
    else:
        return 1


df['Mégafeux'] = df['Surface parcourue (m2)'].apply(assign_value)
```

```python
mgfeux06_21 = df[df['Année']<= 2021].groupby('Année').sum().loc[:,'Mégafeux']
mgfeux06_21.sum()/16
```
